# ごく一部が、全体の半数以上をもっている場合

```r
n <- 10
x <- seq(1, n)
y <- rep(c(1, 10), c((n - 1), 1))
hist(y)
ineq::Lc(y, plot=TRUE)
```

# ごく一部が、全体の2割をもっている場合

```r
n <- 10
x <- seq(1, n)
y <- rep(c(1, 2), c((n - 1), 1))
hist(y)
ineq::Lc(y, plot=TRUE)
```

# 大多数と比べてわずかに多い量を一部が持っている場合

```r
n <- 10
x <- seq(1, n)
y <- rep(c(1, 2), c((n - 2), 2))
hist(y)
ineq::Lc(y, plot=TRUE)
```

# 度数は同じだが、数値に格差があるとき

```r
n <- 10
x <- seq(1, n)
y <- rep(c(1, 2), c((n - 5), 5))
hist(y)
ineq::Lc(y, plot=TRUE)
```

# 度数は同じだが、保有量の格差が大きい場合
